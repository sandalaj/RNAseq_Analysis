# Linear Modeling - R

### Import required packages

```r
library(ggplot2)
```

### Read in the file

```r
read.csv("regrex1.csv")
```

### Assign shorthand for file

```r
regrex1 <- read.csv("regrex1.csv")
```

### Plot the scatterplot

```r
ggplot(regrex1, aes(x,y)) + geom_point()
```

### Plot the scatterplot with the linear model

```r
ggplot(regrex1, aes(x,y)) + geom_point() + geom_smooth(method = "lm", se = F)
```

### Another way of calculating a linear model (no auto graph)

```r
linmod <- lm(y ~ x, data = regrex1)
summary(linmod)
```

### Graphing the linear model using the alternative method above

```r
plot(y ~ x, data = regrex1)
abline(linmod)
```
